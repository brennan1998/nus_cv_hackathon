```python
import os
import re
import cv2
import numpy as np
from os.path import isfile, join
import matplotlib.pyplot as plt
```

```python
col_frames = os.listdir("./nus-sds-dsc2021/train_data/train_images")
```

```python
col_frames
```

```python
# create empty list to store images
col_images = []

for frame in col_frames:
    #read the frames
    temp_img = cv2.imread("./nus-sds-dsc2021/train_data/train_images/" + frame)
    #append the temp image into the list of images
    col_images.append(temp_img)
```

```python
col_images[0].shape
```

```python
num = 20

for frame in [num, num + 1]:
    plt.imshow(cv2.cvtColor(col_images[frame], cv2.COLOR_BGR2RGB))
    plt.title("frame: " + str(frame))
    plt.show()
```

```python
gray = cv2.cvtColor(col_images[frame], cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray,cmap='gray')
```

```python
blurred = cv2.GaussianBlur(gray, (11,11) , 0)
```

```python
plt.imshow(blurred,cmap='gray')
```

```python
canny = cv2.Canny(blurred, 10 , 150, 3)
```

```python
plt.imshow(canny,cmap='gray')
```

```python
dilated = cv2.dilate(canny, (1,1) , iterations = 4)
plt.imshow(dilated, cmap='gray')
```

```python
(cnt,hierachy ) = cv2.findContours(dilated.copy(),cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
```

```python
rgb = cv2.cvtColor(col_images[frame],cv2.COLOR_BGR2RGB)
```

```python
cv2.drawContours(rgb,cnt,-1,(0,255,0),2)
plt.imshow(rgb)
```
